```python
# import the pandas library
import pandas as pd

# import zipfile library (we will use this to extract the file downloaded from Kaggle)
import zipfile

# import kaggle library (we will use this to download the dataset programatically from Kaggle)
import kaggle
```

```python
# Extract Only .csv Files to the Same Directory as the Zip File

import os

# Name of the zip file
zipfile_name = 'amazon-product-dataset-2020.zip'

# Get the directory where the zip file is located
zip_dir = os.path.dirname(os.path.abspath(zipfile_name))

# Open the zip file
with zipfile.ZipFile(zipfile_name, 'r') as file:
    # Iterate through the file names in the zip
    for file_info in file.infolist():
        # Check if the file is a CSV file
        if file_info.filename.endswith('.csv'):
            # Get the file name only (remove any folder structure)
            file_name = os.path.basename(file_info.filename)
            print(file_name)
            
            # Extract the CSV file to the zip file's directory
            with file.open(file_info) as source, open(os.path.join(zip_dir, file_name), 'wb') as target:
                target.write(source.read())
```
